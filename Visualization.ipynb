```python
bb = data.DataReader(name="BB", data_source = "stooq", start = "2007-01-01", end="2020-12-31")
bb.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "Poor"
    elif stock_price <= 50:
        return "Mid"
    else:
        return "Stellar"
    

bb["Low"].apply(rank_performance).value_counts().plot(kind="barh")
```
