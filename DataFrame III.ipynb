```python
# apply()
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_tostring_and_add_millions(number):
    return str(number) + " MILLIONS"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_tostring_and_add_millions)
```

```python
bond.head(3)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Fierce Brosnan":
        return "the best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "No clue"
```

```python
bond.apply(good_movie, axis ="columns")
```
